```python
df = df.dropna()
```

```python
df["Category"]= df["Category"].str.strip().str.upper()
df["City"]= df["City"].str.strip().str.upper()
```

```python
df
```

```python
df["UnitPrice"]=df["UnitPrice"].replace(r'[\$]', '', regex=True)
```

```python
df
```

```python
df["UnitPrice"] = df["UnitPrice"].replace(r"[^\d.]", "", regex=True).astype(float)
df["Revenue"] = df["Revenue"].replace(r"[^\d.]", "", regex=True).astype(float)
```

```python
df
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"], format= "mixed")
```
